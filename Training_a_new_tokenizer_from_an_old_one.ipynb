```python
#importing the necessary libraries
from datasets import load_dataset
```

```python
#loading in the datasets
datasets = load_dataset("code_search_net", "python")
datasets
```

```python
#checking a sample
print(datasets['train'][345678]["whole_func_string"])
```

```python
#transforming the datasets into an iterator list of object

training_corpus = (
    datasets["train"][i: i+1000]["whole_func_string"]
    for i in range(0, len(datasets["train"]),1000)
)
```

```python
#defining a function to return the generator

def get_training_corpus():
  return (
      datasets["train"][i:i+1000]["whole_func_string"]
      for i in range(0, len(datasets["train"]), 1000)
  )
```

**Training a new tokenizer**

```python
#loading the old tokenizer

from transformers import AutoTokenizer
old_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
#testing how the old tokenizer perform on one of the sample
example = datasets["train"][345678]["whole_func_string"]
tokens = old_tokenizer.tokenize(example)
print(tokens)
```

```python
#training the new tokenizer

tokenizer = old_tokenizer.train_new_from_iterator(training_corpus, 52000)
```

```python
#testing the newly trained tokenizer

tokens = tokenizer.tokenize(example)
print(tokens)
```

```python
#Saving the tokenizer
tokenizer.save_pretrained("code-search-net-tokenizer")
```
